```python
# Split the dataset into X (features) and y (target variable)
X = df.drop('HeartDiseaseorAttack', axis=1)
y = df['HeartDiseaseorAttack']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the Seaborn style and increase the border size
sns.set(style='white', rc={'axes.linewidth': 0})

# Create a custom color palette with darker shades of blue
my_palette = sns.color_palette(["#3EDAD8","#01A0C4"])

# Convert the resampled data to a DataFrame
df = pd.concat([X, y], axis=1)

# Plot a countplot of the HeartDiseaseorAttack variable with the custom color palette
ax = sns.countplot(data=df, x='HeartDiseaseorAttack', palette=my_palette)

# Add the exact number on each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.0f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

# Show the plot
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
# Apply SMOTE to balance the data
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
# Print the shape of the original and resampled data
print('Original dataset shape:', X.shape, y.shape)
print('Resampled dataset shape:', X_resampled.shape, y_resampled.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the Seaborn style and increase the border size
sns.set(style='white', rc={'axes.linewidth': 0})

# Create a custom color palette with darker shades of blue
my_palette = sns.color_palette(["#3EDAD8","#01A0C4"])

# Convert the resampled data to a DataFrame
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)

# Plot a countplot of the HeartDiseaseorAttack variable with the custom color palette
ax = sns.countplot(data=df_resampled, x='HeartDiseaseorAttack', palette=my_palette)

# Add the exact number on each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.0f}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

# Show the plot
plt.show()
```

```python
# Plot a correlation matrix using a heatmap
corr_matrix = df_resampled.corr()

# Create a table of the correlation matrix
corr_table = corr_matrix.style.background_gradient(cmap='coolwarm')

# Display the table
corr_table
```

```python
# Select the target variable
target_variable = 'HeartDiseaseorAttack'

# Rank the variables based on their correlation with the target variable
corr_rank = corr_matrix[target_variable].sort_values(ascending=False)
""
# Select the top 12 variables with the highest correlation coefficients
top12_corr = corr_rank[1:13]
last9_corr = corr_rank[13:22]

# Print the top 12 correlated variables and the 9 other variables also which are less corelated
print("Top 12 correlated variables in asc order: \n", top12_corr)
print("Remaining 9 correlated variables in asc order: \n", last9_corr)
```

```python
# Create a new DataFrame with the target variable and the top 12 features
df = df_resampled[list(top12_corr.index)+[target_variable]]
```

```python
from sklearn.model_selection import train_test_split

# Set the target variable to "HeartDiseaseorAttack"
target = 'HeartDiseaseorAttack'
X = df.drop(columns=[target])
y = df[target]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
# Initialize StandardScaler object
scaler = StandardScaler()

# Scale the training data
X_train_scaled = scaler.fit_transform(X_train)

# Scale the test data using the same scaler object
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error

# Initialize Decision Tree Classifier with default parameters
dt = DecisionTreeClassifier()

# Train the model on the scaled training data
dt.fit(X_train_scaled, y_train)

# Predict on the scaled test data
y_pred = dt.predict(X_test_scaled)

# Calculate the confusion matrix, accuracy and classification report of the model
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
cr = classification_report(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print('Decision Tree: \n')
print('Confusion Matrix: \n', cm)
print('Accuracy: ', accuracy)
print('Mean Squared Error: ', mse)
print('Classification Report: \n', cr)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, mean_squared_error

# Initialize the AdaBoost Classifier
ada = AdaBoostClassifier(random_state=42)

# Train the model on the scaled training data
ada.fit(X_train_scaled, y_train)

# Predict on the scaled test data
y_pred = ada.predict(X_test_scaled)

# Calculate the confusion matrix, accuracy, MSE and classification report of the model
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
cr = classification_report(y_test, y_pred)

# Print the results
print('AdaBoost: \n')
print('Confusion Matrix: \n', cm)
print('Accuracy: ', accuracy)
print('Mean Squared Error: ', mse)
print('Classification Report: \n', cr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()

rfc.fit(X_train_scaled, y_train)

y_pred = rfc.predict(X_test_scaled)

# Calculate the confusion matrix, accuracy, MSE and classification report of the model
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
cr = classification_report(y_test, y_pred)

# Print the results
print('Random Forrest: \n')
print('Confusion Matrix: \n', cm)
print('Accuracy: ', accuracy)
print('Mean Squared Error: ', mse)
print('Classification Report: \n', cr)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# create a logistic regression object
lr = LogisticRegression(max_iter=1000)

# fit the logistic regression model to the oversampled data
lr.fit(X_train_scaled, y_train)

# predict the target variable using logistic regression
y_pred_lr = lr.predict(X_test_scaled)

cm = confusion_matrix(y_test, y_pred_lr)

# generate classification report for logistic regression
report_lr = classification_report(y_test, y_pred_lr)

# calculate accuracy score for logistic regression
acc_score_lr = accuracy_score(y_test, y_pred_lr)

#calculate mean squared error
mse = mean_squared_error(y_test, y_pred_lr)

# print the results for logistic regression
print("Logistic Regression: \n")
print('Confusion Matrix: \n', cm)
print("Accuracy Score:", acc_score_lr)
print("Mean Squared Error:", mse)
print("Classification Report:")
print(report_lr)
```

```python
from sklearn.naive_bayes import GaussianNB

# Build the Naive Bayes model
nb = GaussianNB()
nb.fit(X_train_scaled, y_train)

# Evaluate the model on the test set
y_pred = nb.predict(X_test_scaled)

# Calculate evaluation metrics
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
cr = classification_report(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

# Print the results
print('Naive Bayes: \n')
print('Confusion Matrix: \n', cm)
print('Accuracy: ', accuracy)
print('Mean Squared Error: ', mse)
print('Classification Report: \n', cr)
```

```python
# use ANN to predict the HeartDiseaseorAttack
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import auc

# build a ANN model to predict the HeartDiseaseorAttack
mlp = MLPClassifier(hidden_layer_sizes=(40,40,40))
mlp.fit(X_train_scaled, y_train)
# predict the HeartDiseaseorAttack
y_pred = mlp.predict(X_test_scaled)

# Calculate evaluation metrics
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
cr = classification_report(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

# Print the results
print('ANN: \n')
print('Confusion Matrix: \n', cm)
print('Accuracy: ', accuracy)
print('Mean Squared Error: ', mse)
print('Classification Report: \n', cr)
```

```python
# Evaluation Metrics Comparison-1

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# create a sample dataframe with random values
data = {'Metric': ['Accuracy', 'Precision', 'Recall'],
         'Random Forrest': [0.92, 0.92, 0.92],
        'Logistic Regression': [0.77, 0.77, 0.77],
        'KNN': [0.87, 0.89, 0.85],
        'Naive Bayes': [0.74, 0.73, 0.76]}
df = pd.DataFrame(data)
df.set_index('Metric', inplace=True)

# define a color palette of blue shades
colors = ['#3C5F8D', '#4872A3', '#5686B8']

# create four bar plots
fig, axs = plt.subplots(ncols=4, figsize=(12, 4))
for i, col in enumerate(df.columns):
    bars = axs[i].bar(df.index, df[col], color=colors)
    axs[i].set_title(col)
    axs[i].set_ylim([0, 1])
    axs[i].set_ylabel('Metric Value')
    axs[i].tick_params(axis='x', rotation=45)
    for bar in bars:
        height = bar.get_height()
        axs[i].text(bar.get_x() + bar.get_width() / 2.0, height, f'{height:.2%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
